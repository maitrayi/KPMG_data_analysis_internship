```python
import pandas as pd
import numpy as np
import os 
from datetime import datetime
```

```python
pwd = os.getcwd()
file_path= os.path.join(pwd ,'KPMG.xlsx')
data = pd.read_excel(file_path, sheet_name='NewCustomerList')
```

```python
data=data.dropna()
```

```python
data['DOB'] = pd.to_datetime(data['DOB'])
```

```python
current_year = datetime.now().year
data['Age'] = current_year - data['DOB'].dt.year
```

```python
data['index'] = data.index
```

```python
data = data.drop(['DOB', 'job_title', 'deceased_indicator', 'address', 'postcode', 'country', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',
       'Unnamed: 19', 'Unnamed: 20','Rank', 'Value',], axis=1)
```

```python
data.to_excel('new_customer_data.xlsx', index=False)
```
