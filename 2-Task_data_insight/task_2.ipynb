```python
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt
import plotly.express as ps 
from datetime import datetime
```

```python
transaction = 'Transactions'
demographic = 'CustomerDemographic'
address = 'CustomerAddress'
new_customer = 'NewCustomerList'
```

```python
pwd = os.getcwd()
file_path= os.path.join(pwd ,'KPMG.xlsx')
transaction_data_set = pd.read_excel(file_path, sheet_name= transaction)
demographic_data_set = pd.read_excel(file_path, sheet_name= demographic)
address_data_set = pd.read_excel(file_path, sheet_name= address)
new_customer_data_set = pd.read_excel(file_path, sheet_name= new_customer)
```

```python
demographic_data_set['DOB'] = pd.to_datetime(demographic_data_set['DOB'])
current_date = datetime.now()
demographic_data_set= demographic_data_set.dropna()
demographic_data_set['Age'] = current_date.year -demographic_data_set['DOB'].dt.year
```

```python
transaction_data_set= transaction_data_set.dropna()
```

```python
transaction_data_set['Profit']= transaction_data_set['list_price'] - transaction_data_set['standard_cost']
```
